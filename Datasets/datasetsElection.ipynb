Las enfermedades y los problemas en el corazón afectan a una gran parte de la sociedad. Nos afectan a nosotros ya sea de forma directa o a nuestros familiares o amigos. Este dataset pertenece al BRFSS 2015 (Behavioral Risk Factor Surveillance System). Se busca predecir si un individuo presenta problemas en el corazón teniendo en cuentas distintas características nombradas mas abajo. Se busca utilizar una clasificación binaria.

```python
# This is a Heart Desease dataset
url = 'https://raw.githubusercontent.com/Tomtechno/DataScience-Coder/main/heart_disease_health_indicators_BRFSS2015.csv?token=GHSAT0AAAAAABZIXNY35I43NJ2BCJG2S7EYY25QKGQ'
df3 = pd.read_csv(url)
df3.head()
```

```python
print('# Rows:    ',df3.shape[0])
print('# Columns: ',df3.shape[1])
df3.info()
```

Considero importante a los atributos:


*   HeartDiseaseorAttack - 1 si el individuo tiene problemas al corazón. 0 no tiene problemas al corazón.
*   HighBP - 1 si el individuo tiene presion sanguínea alta. 0 no tiene presion sanguínea alta.
*   HighChol - 1 si el individuo tiene colesterol. 0 si no tiene colesterol.
*   CholCheck - 1 si el individuo tiene un chequeo de colesterol en los ultimos 5 anios. 0 si no tiene chequeo.
*   BMI - Lugar de residencia (rural o ciudad).
*   Smoker - 1 si el individuo fumo al menos 100 cigarrillos. 0 si no lo hizo.
*   Stroke - 1 si el individuo fue diagnosticado con dolores de cabeza. 0 si no fue diagnosticado.
*   Diabetes - 1 si el individuo fue diagnosticado con diabetes. 0 si no fue diagnosticado.
*   PhysActivity - 1 si el individuo hizo actividad física en los ultimos 30 dias. 0 si no hizo.
*   Fruits - 1 si el individuo consume al menos 1 fruta por dia. 0 si no lo hace.
*   Veggies - 1 si el individuo consume al menos 1 verdura por dia. 0 si no lo hace.
*   HvyAlcoholConsump - 1 si es hombre y consume mas de 14 vasos de alcohol por semana o si es mujer y consume mas de 7 vasos de alcohol por semana. 0 sino lo hacen.
*   GenHlth - Nivel de actividad fisica.
*   MentHlth - Cantidad de dias con problemas mentales en este ultimo mes.
*   PhysHlth - Cantidad de dias con problemas físicos en este ultimo mes.
*   DiffWalk - 1 si tiene problemas al caminar. 0 si no.
*   Sex - Nivel medio de glucosa en sangre.
*   Age - Nivel medio de glucosa en sangre.
*   Education - Nivel medio de glucosa en sangre.
*   Income - Nivel medio de glucosa en sangre.
